```python
import pandas as pd
```

```python
def merge(path, fileName):
    trainName = '../models/__models__/{}_{}.csv'.format(path, fileName)
    testName = trainName.replace('train', 'test')
    train = pd.read_csv(trainName)
    test = pd.read_csv(testName)
    data = pd.concat([train, test])
    return data
```

```python
data = pd.read_csv('../input/data.csv')

data_stacking = data[['Id', 'Score']].copy()
data_stacking.head()
```

```python
models = [100, 101, 102]
for model in models:
    data_stacking = pd.merge(data_stacking, merge('lgbm_train', model), on = 'Id', how = 'left')
data_stacking.head()
```

```python
models = [1, 2, 3, 4, 5]
for model in models:
    data_stacking = pd.merge(data_stacking, merge('train_lgbmc_lc_label', model), on = 'Id', how = 'left')
data_stacking.head()
```

```python
data_stacking = pd.merge(data_stacking, merge('train_clf_0_label_lc_label', 1), on = 'Id', how = 'left')
data_stacking.head()
```

```python
models = [1, 2, 3, 4, 5]
clfs = [0, 1, 2, 3]
for clf in clfs:
    for model in models:
        data_stacking = pd.merge(data_stacking, merge('train_clf_{}_label_lc_label'.format(clf), model), on = 'Id', how = 'left')
data_stacking.head()
```

```python
data_stacking = pd.merge(data_stacking, merge('train_lstm_len', 10), on = 'Id', how = 'left')
data_stacking.head()
```
