```python
import pandas as pd
import numpy as np
from sklearn import metrics
```

```python
def rmse(y_train, y_pred):
    return np.sqrt(metrics.mean_squared_error(y_train, y_pred))
```

```python
train_result = pd.read_csv('../result/train_0.6056817644422253.csv')

# find 4
maxv = 0
best_threshold = 0
for threshold in range(4000, 5001):
    threshold = threshold / 1000
    train = train_result.copy()
    train['Predict'].ix[(train['Predict'] >= 4.0) & (train['Predict'] < threshold)] = 4
    train['Predict'].ix[(train['Predict'] >= threshold)] = 5
    error = rmse(train['Predict'].values, train['Score'].values)
    error = 1 / (1 + error)
    if error > maxv:
        best_threshold = threshold
        maxv = error
        print(maxv, threshold)
```
